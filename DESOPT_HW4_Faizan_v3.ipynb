```python
import numpy 
import matplotlib.pyplot as plt

# define grid
X1 = numpy.linspace(-2, 3.5, 101)
X2 = numpy.linspace(3, -2, 101)
x1, x2 = numpy.meshgrid(X1, X2)

#define objective function
obj_f = numpy.array( [(x1+1) **2 + (x2-2) **2 for x2 in X2 for x1 in X1]).reshape(x1.shape)
```

```python
# plot objective function in contour plot
plt.figure(1)
contour = plt.contour(x1, x2, obj_f, cmap=plt.cm.Spectral, levels=[0, 3, 6, 9, 12, 15, 18, 21])
plt.colorbar(contour, ticks=[0, 3, 6, 9, 12, 15, 18, 21])
plt.xlabel('x1')
plt.ylabel('x2')
# plotting gradients at corner points

# @corner point (0, 0)
plt.text(-0.5, 0.25, r'$\frac{\partial g_3}{\partial x}$', fontsize=18)
plt.arrow(0, 0, 0, -0.5)
plt.text(-0.55, -0.75, r'$\frac{\partial g_4}{\partial x}$', fontsize=18)
plt.arrow(0, 0, 0.25, -0.5)
plt.text(0.35, -0.5, r'$\frac{\partial f}{\partial x}$', fontsize=18)
plt.arrow(0, 0, -0.5, 0)


# @corner point (2, 0)
plt.text(2, 0.2, r'$\frac{\partial g_1}{\partial x}$', fontsize=18)
plt.arrow(2, 0, 0, -0.5)
plt.text(1.5, -0.5, r'$\frac{\partial g_4}{\partial x}$', fontsize=18)
plt.arrow(2, 0, 0.37, -0.25)
plt.text(2.3, -0.6, r'$\frac{\partial f}{\partial x}$', fontsize=18)
plt.arrow(2, 0, 0.5, 0)

# @corner point (2, 1)
plt.text(2.4, 1.3, r'$\frac{\partial g_1}{\partial x}$', fontsize=18)
plt.arrow(2, 1, 0, 0.5, width=0.02)
plt.text(1.5, 1.5, r'$\frac{\partial g_2}{\partial x}$', fontsize=18)
plt.arrow(2, 1, 0.37, -0.125)
plt.text(2.5, 0.55, r'$\frac{\partial f}{\partial x}$', fontsize=18)
plt.arrow(2, 1, 0.5, 0, width=0.02)


# @corner point (0, 1)
plt.text(0, 1.6, r'$\frac{\partial g_2}{\partial x}$', fontsize=18)
plt.arrow(0, 1, 0.25, -0.25)
plt.text(-0.6, 1.2, r'$\frac{\partial g_3}{\partial x}$', fontsize=18)
plt.arrow(0, 1, 0, 0.5)
plt.text(0.25, 0.6, r'$\frac{\partial f}{\partial x}$', fontsize=18)
plt.arrow(0, 1, -0.5, 0)

# plot given constraints and feasible domain
plt.plot(X1, numpy.zeros(X1.shape[0]), c='black', linewidth=0.5)
plt.plot(X1, numpy.ones(X1.shape[0])*1, c='black', linewidth=0.5)
plt.plot(numpy.zeros(X1.shape[0]), X2, color='black', linewidth=0.5)
plt.plot(numpy.ones(X1.shape[0])*2, X2, color='black', linewidth=0.5)
plt.fill([0, 2, 2, 0], [0, 0, 1, 1], "g", alpha=0.5, label='Feasible domain')
plt.legend(loc='upper left')

plt.show()
```

## Problem 2 Solution Code

```python
import numpy
import matplotlib.pyplot as plt

#define grid
x1 = numpy.linspace(-2, 1.5, 101)
x_domain = numpy.linspace(-2, 1.5, 91)
x2 = (1-x_domain) ** 3
```

```python
# plot figure and constraints
plt.figure(2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.plot(x1, numpy.zeros(x1.shape[0]), c='black', linewidth=0.5)
plt.plot(numpy.zeros(x2.shape[0]),x2, c='black', linewidth=0.5)
plt.fill_between(x_domain, x2, color='green', alpha=0.5, label='Feasible domain')
plt.scatter(1, 0, marker='o', s=20, c='k')
plt.legend(loc='upper right')
plt.ylim(-10, 30)

# plot the gradient at corner points
# @corner point (1, 0)
plt.arrow(1, 0, -0.5, 0, head_width=1, head_length=0.05, length_includes_head=True)
plt.text(0.5, 1.8, r'$\frac{\partial f}{\partial x}$', fontsize=18)
plt.arrow(1, 0, 0, 4, head_width=0.05, head_length=0.75, length_includes_head=True)
plt.text(0.9, 6, r'$\frac{\partial g_1}{\partial x}$', fontsize=18)
plt.arrow(1, 0, 0, -4, head_width=0.05, head_length=0.75, length_includes_head=True)
plt.text(0.9, -8, r'$\frac{\partial g_2}{\partial x}$', fontsize=18)

plt.show()
```
